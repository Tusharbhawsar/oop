```python
def insert_patien_data(name:str,age:int):

    if type(name)==str and type(age)==int:
        print(name)
        print(age)
        print("inserted into database")
    else:
        raise TypeError("incorrect data type")

insert_patien_data("tushar",30)  #this code 
```

```python
def insert_patien_data(name:str,age:int):

    if type(name)==str and type(age)==int:
        if age < 0:
            raise TypeError("age not less than 0")
        else:
            print(name)
            print(age)
            print("inserted into database")
    else:
        raise TypeError("incorrect data type")

insert_patien_data("tushar",-1)  #this code 
```
