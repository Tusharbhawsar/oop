```python
# protected modifier

class Demo:
    def __init__(self):
        self._email = "tushar@gmail.com"

    def return_main(self):
        return self._email
    
obj = Demo()
# print(dir(obj))
# obj._email
obj.return_main()
```

```python
#  private modifier

class Demo2:
    def __init__(self):
        self.__address = "akhand nagar"

    def __return_add(self):
        return self.__address
    
obj2 = Demo2()
# obj2.__address    #not accesible
# obj2.Demo2_address
# obj2._Demo2__address()
obj2._Demo2__return_add()
```

```python
class Password():
    def __init__(self):
        self.__password = "Admin123"
    
    def __passa(self):
        return self.__password
        
        
obj = Password()
print(obj._Password__passa())
```
