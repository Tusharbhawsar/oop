```python
from abc import ABC,abstractmethod


class BankApp(ABC):

    def database(self):
        print("connected to database")

    @abstractmethod
    def security(self):
        pass

    @abstractmethod
    def display(self):
        pass


class MobileApp(BankApp):

    def mobile_login(self):
        print("Login into mobile")

    def security(self):
        print("mobile security")

    def display(self):
        print("display security")
    
```

```python
mobile = MobileApp()
mobile.security()
mobile.display()
```
