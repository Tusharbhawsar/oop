```python
class Parent:
    def __init__(self, num):
        self.__num = num

    def get_num(self):
        return self.__num


class Child(Parent):
    def __init__(self, val, num):
        super().__init__(num)   # Call Parent constructor
        self.__val = val        # Correct attribute name

    def get_val(self):
        return self.__val


s = Child(100, 2)
print("parent:NUM", s.get_num())
print("child:VAL", s.get_val())


# agar child class ke pass apna constructor nhi h to parent ka constructo autimatically invoke ho jata h
# but agar child class ke pass apna constructor h to parent ka constructor call nhi hoga unless we explicitly call it using super()
```
