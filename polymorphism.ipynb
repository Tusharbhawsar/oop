```python
class Geometry:
    def area(self, *args):
        if len(args) == 1:
            a = args[0]
            print("Circle", 3.14 * a * a)

        elif len(args) == 2:
            a, b = args
            print("Rect", a * b)

        else:
            print("Invalid number of arguments")

obj = Geometry()
obj.area(4)      # circle
obj.area(4, 5)   # rectangle
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # Overloading +
    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)

p1 = Point(3, 4)
p2 = Point(1, 2)
p3 = p1 + p2   # operator overloading in action

print(p3.x, p3.y)
```

```python
import requests

class Product:
    URL = "https://api.freeapi.app/api/v1/public/randomproducts/product/random"

    def __init__(self):
        # fetch once and reuse
        self._data = self._fetch_data()

        user_input = input(
            "How would you like to proceed?\n"
            "1. Enter 1 to check the brand\n"
            "2. Enter 2 to check the price\n"
            "3. Enter 3 to check the category\n"
            ">>> "
        )

        if user_input == "1":
            print(self.brand())
        elif user_input == "2":
            print(self.price())
        elif user_input == "3":
            print(self.category())
        else:
            print("Invalid option selected.")

    def _fetch_data(self):
        response = requests.get(self.URL)
        response.raise_for_status()
        return response.json()["data"]

    def brand(self):
        return self._data.get("brand")

    def price(self):
        return self._data.get("price")

    def category(self):
        return self._data.get("category")


# Create object and call a function
prod = Product()
print(prod.price())
print(prod.brand())

```
